```
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```
book = pd.read_csv('book.csv')
book
```

##Apriori Algorithm

####With Min_Support 0.18

```
freq_itemsets = apriori(book,min_support=0.18,use_colnames=True)
freq_itemsets
```

####With Min_Support=0.15

```
freq_itemsets1 = apriori(book,min_support=0.15,use_colnames=True)
freq_itemsets1
```

####With Min_Support 0.1

```
freq_itemsets2 = apriori(book, min_support=0.1,use_colnames=True)
freq_itemsets2
```

#Association Rules

###Using 0.18 Min_Support and Metric='Lift'


```
rules = association_rules(freq_itemsets,metric='lift',min_threshold=0.7)
rules
```

###Using 0.18 Min_Support and Metric='Confidence'

```
rules1 = association_rules(freq_itemsets,metric='confidence',min_threshold=0.6)
rules1
```

###Using 0.15 Min_Support and Metric='Lift'


```
rules_1 = association_rules(freq_itemsets1,metric='lift',min_threshold=0.7)
rules_1
```

```
rules_2 = association_rules(freq_itemsets1,metric='confidence',min_threshold=0.7)
rules_2
```

###Using 0.15 Min_Support and Metric='lift'

```
rules_3 = association_rules(freq_itemsets2,metric='lift',min_threshold=1)
rules_3
```

###Using 0.15 Min_Support and Metric='Confidence'

```
rules_4 = association_rules(freq_itemsets2,metric='confidence',min_threshold=0.7)
rules_4
```

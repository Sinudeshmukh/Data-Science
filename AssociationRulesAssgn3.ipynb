```
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```
df = pd.read_csv('my_movies.csv')
df
```

```
df.drop(df.columns[[0,1,2,3,4]],axis=1,inplace=True)
```

```
df
```

#Apriori Algorithm

###With Min_Support=0.4

```
freq_itemsets = apriori(df,min_support=0.4,use_colnames=True)
freq_itemsets
```

###With Min_Support=0.2

```
freq_itemsets1 = apriori(df,min_support=0.2,use_colnames=True)
freq_itemsets1
```

#Association Rules

###Using Metric Lift and Min_Support=0.4

```
rules = association_rules(freq_itemsets,metric='lift',min_threshold=0.7)
rules
```

###Using Metric Confidence and Min_Support=0.4

```
rules1 = association_rules(freq_itemsets,metric='confidence',min_threshold=0.7)
rules1
```

###Using Metric Confidence and Min_Support=0.2

```
rules_1 = association_rules(freq_itemsets1,metric='confidence',min_threshold=0.7)
rules_1
```

###Using Metric Lift and Min_Support=0.2

```
rules_2 = association_rules(freq_itemsets1,metric='lift',min_threshold=0.7)
rules_2
```

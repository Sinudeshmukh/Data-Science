```
  import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```
df=pd.read_csv("bank-full.csv",sep=";")
```

```
df.head()
```

```
df_1= df.drop(['month','day'],axis=1)
```

```
df_1=pd.get_dummies(data=df_1, columns=['job','marital','education',"default",'housing','loan','contact','poutcome','y'])
```

```
df_1
```

```
df_1=df_1.drop(['y_no'],axis=1)
```

```
df_1= df_1.rename({'y_yes':'Subscribed'},axis=1)
```

```
df_1.corr()
```

```
df_1.shape
```

```
df_1.info()
```

```
X = df_1.iloc[:,0:38]
Y = df_1.iloc[:,38]
```

```
from sklearn.linear_model import LogisticRegression
```

```
classifier = LogisticRegression()
classifier.fit(X,Y)
```

```
y_pred = classifier.predict_proba(X)
```

```
y_pred
```

```
y_pred = classifier.predict(X)
```

```
y_pred
```

```
Y_Predicted = pd.DataFrame({'actual':Y,
                            'Predicted':y_pred})
```

```
Y_Predicted
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix=confusion_matrix(Y,y_pred)
confusion_matrix
```

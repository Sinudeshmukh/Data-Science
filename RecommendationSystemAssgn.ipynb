```
import pandas as pd
import numpy as np
```

```
path = 'book.csv'
df = pd.read_csv(path,sep=None)
```

```
df
```

```
df = df.drop(['Unnamed: 0'],axis=1)
```

```
df = df.rename(columns={'User.ID':'user_id','Book.Title':'book_title','Book.Rating':'book_rating'})
```

```
df
```

```
len(df.user_id.unique())
```

```
len(df.book_title.unique())
```

```
df['book_rating'].value_counts()
```

```
len(df.book_rating.unique())
```

```
df = df.drop_duplicates(['user_id','book_title'])
```

##Pivot table for better Visualization 

```
books = df.pivot(index='user_id',
                  columns='book_title',
                  values='book_rating',).reset_index(drop=True)
```

```
books
```

```
books.index = df.user_id.unique()
```

```
books
```

```
books.fillna(value=0,inplace=True)
```

```
books
```

#Creating Cosine Similarity Between Users

```
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```
user_sim = 1 - pairwise_distances(books.values,metric='cosine')
```

```
user_sim
```

```
user_sim_df = pd.DataFrame(user_sim)
```

```
user_sim_df.index
```

```
user_sim_df.index = df.user_id.unique()
user_sim_df.columns = df.user_id.unique()
```

```
user_sim_df.iloc[0:5,0:5]
```

```
np.fill_diagonal(user_sim,0)
user_sim_df
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.stripplot(y='tip',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='tip',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='tip',data=tips,jitter=True,hue='sex')
```

```python
sns.stripplot(x='day',y='tip',data=tips,jitter=False)
plt.xlabel('DAYS')
plt.ylabel('TIPS $')
plt.show()
```

```python
sns.swarmplot(data=tips,x='day',y='tip')
```

```python
sns.swarmplot(data=tips,x='day',y='tip',hue='sex')
plt.title('TIPS')
plt.show()
```

```python
sns.swarmplot(data=tips,x='tip',y='day',hue='sex',orient='h')
```
